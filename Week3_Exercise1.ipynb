# EXERCISE 1: Data Analysis using mtcars

```python
import pandas as pd
df=pd.read_csv("mtcars.csv")
df
```

# Q-1 Find the car with the best mpg

```python
df1=df[['model']][df.mpg==df['mpg'].max()]
df1
```

# Q-2 Find the car with the worst mpg

```python
df2=df[['model']][df.mpg==df['mpg'].min()]
df2
```

# Q-3 Find the car with the best horsepower

```python
df3=df[['model']][df.hp==df['hp'].max()]
df3
```

# Q-4 Find 5 number summary of displacement

```python
df['disp'].describe()
```

# Q-5 Find median horse power

```python
df['hp'].median()
```

# Q-6 What is average mpg for manual vs. automatic cars

```python
df1=df[df.am==1]
df2=df[df.am==0]
print("The avg of manual cars is "+str(df1.mpg.mean()))
print("The avg of automatic cars is "+str(df2.mpg.mean()))
```
