# EXERCISE 1: Data Analysis using CEREALS dataset
Use the prepared or preprocessed data to answer the following:

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_excel("Cereals1.xls")
```

```python
df
```

```python
df=df.mask(df==-1).fillna(df.mean())
```

```python
df
```

```python
column=["protein", "fat", "sodium", "fiber", "carbo", "sugars", "potass", "vitamins"]
for i in column:
    Q1=np.quantile(df[i],0.25)
    Q3=np.quantile(df[i],0.75)
    Q=np.quantile(df[i],0.50)
    IQR= Q3-Q1
    upper= Q3+1.5*IQR
    lower= Q1-1.5*IQR
    df[i] = df[i].apply(lambda x: Q if x < lower or x > upper else x)
df    
```

# Q-1 Cross tabulate the type of cereal (hot vs cold) against the manufacturer

```python
pd.crosstab(df['type'],df['mfr'])
```

# Q-2 Which is the cereal with the best rating, worst rating?

```python
best=df[['name']][df.rating==df['rating'].max()]
best
```
