```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Roll Number' : [101,102,103,104,105,106,107,108,109,110],
    'Name' : ['Alice', 'Bob', 'Eve', 'Raj','charlie', 'David', 'Nihal', 'Krish', 'Anne', 'Vans'],
    'Gender' : ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks1' : [79, 90, 69, 56, 89, 60, 90, 47, 34, 32],
    'Marks2' : [70, 90, 80, 50, 37, 26, 78, 42, 78, 90],
    'Marks3' : [67, 40, 89, 96, 92, 95, 80, 57, 67, 70]
}
df = pd.DataFrame(data)
df
```

```python
df['total marks'] = df['Marks1']+df['Marks2']+df['Marks3']
df
```

```python
lowest_marks = df['Marks1'].min()
print('Lowest marks in Marks1',lowest_marks)
```

```python
highest_marks = df['Marks2'].max()
print('Highest marks in Marks2',highest_marks)
```

```python
avg_marks = df['Marks3'].mean()
print('Average marks of Marks3',avg_marks)
```

```python
df['Average Marks'] = df[['Marks1', 'Marks2', 'Marks3']].mean(axis=1)
df
```

```python
stu_hi_avg = df[df['Average Marks']==df['Average Marks'].max()]['Name'].values[0]
print("Student with highest average marks",stu_hi_avg)
```
