# Q-4 For each of the numeric attributes (proteins up to vitamins), identify and replace all noisy data with the median of attribute

```python
column=["protein", "fat", "sodium", "fiber", "carbo", "sugars", "potass", "vitamins"]
for i in column:
    Q1=np.quantile(df[i],0.25)
    Q3=np.quantile(df[i],0.75)
    Q=np.quantile(df[i],0.50)
    IQR= Q3-Q1
    upper= Q3+1.5*IQR
    lower= Q1-1.5*IQR
    df[i] = df[i].apply(lambda x: Q if x < lower or x > upper else x)
df    
```
